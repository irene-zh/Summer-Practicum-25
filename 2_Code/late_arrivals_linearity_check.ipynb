```python
import numpy as np
import pandas as pd
import os
from plotnine import *
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap, TwoSlopeNorm
```

```python
path = r'D:\Summer Practicum\2_Code\arrdelayblk_grouped_avg_filt.csv'
df = pd.read_csv(path)
```

```python
plt.rcParams.update({'font.size':10})
sns.scatterplot(data = df, x = "ArrDelayBlk", y = "ANext_5", color='#9C2C19', s = 10)
plt.title("Arrival Delays vs. 5 Consequent Arrival Delays")
plt.xlabel("Arrival Delay (Quantized)")
plt.ylabel("Average Delay of 5 Consecutive Arrivals")
plt.gca().set_aspect('equal', adjustable='datalim')  # or 'datalim'

# Get current axis limits to draw y = x across full range
xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()

# Define the line range
line_min = min(xmin, ymin)
line_max = max(xmax, ymax)
plt.xlim(-60, 180)
plt.ylim(-60, 180)  # now Y range matches X range
plt.plot([line_min, line_max], [line_min, line_max], color='gray', linestyle='--', label='y = x')

plt.show()
```
